```python
import pandas as pd
```

```python
data = pd.read_csv('/content/finalcsv.csv')
data.head()
```

```python
# Group utterances by intent
grouped_utterances = data.groupby('intent')['utterance'].apply(list)

for intent, utterances in grouped_utterances.items():
    print(f"\nIntent: {intent}")
    print("Utterances:")
    for i, utterance in enumerate(utterances, 1):
        print(f"  {i}. {utterance}")
```

```python
import pandas as pd
import yaml

# Assuming 'data' is your DataFrame with 'utterance' and 'intent' columns
grouped_data = data.groupby('intent')['utterance'].apply(list).to_dict()

# Format the data in YAML structure for NLU
nlu_data = {
    "version": "3.1",
    "nlu": []
}

for intent, examples in grouped_data.items():
    intent_data = {
        "intent": intent,
        "examples": []
    }

    for example in examples:
        intent_data["examples"].append(f"{example}")

    nlu_data["nlu"].append(intent_data)

# Save to YAML file
with open("nlu_data.yml", "w") as yaml_file:
    yaml.dump(nlu_data, yaml_file, default_flow_style=False, sort_keys=False)

print("Data successfully saved to nlu_data.yml")
```
